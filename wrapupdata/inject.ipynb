```python
from models.diffuserpipeline import StableDiffusionPipelineForNegativePrompts
import torch
from daam.utils import set_seed
from matplotlib import pyplot as plt
model_id = 'stabilityai/stable-diffusion-2-base'
device = 'cuda'
pipe = StableDiffusionPipelineForNegativePrompts.from_pretrained(model_id, use_auth_token=True)
pipe = pipe.to(device)

prompt = "professional office woman"
negative_prompt = "glasses"
seed = 8577767
steps = 30
words = "n:glasses"

with torch.cuda.amp.autocast(dtype=torch.float16), torch.no_grad():
    out = pipe.inject(prompt, negative_prompt=negative_prompt, num_inference_steps=steps, generator=set_seed(seed),inject_step=[0],inject_guidance = 10)
    
plt.imshow(out.images[0])
```
