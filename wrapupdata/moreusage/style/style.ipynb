```python
# mean activation value of the feature maps
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "0"
import argparse
from diffusers import StableDiffusionPipeline
import torch
from daam import set_seed
import matplotlib.pyplot as plt

model_id = 'stabilityai/stable-diffusion-2-base'
device = 'cuda'
pipe = StableDiffusionPipeline.from_pretrained(model_id, use_auth_token=True)
pipe = pipe.to(device)

prompt = "an oil painting of sunflowers"
negative_prompt = "impressionism"

fig,axs = plt.subplots(2, 2, figsize=(10, 10))
axs = axs.flatten()
for ax in axs:
    ax.axis('off')
fig.tight_layout
negative_times = [-1, 5, 15, 31]
titles = ['Negative prompts 0~30', 'negative prompts 5~30', 'negative time = 15~30', 'No negative prompts']
seed = 10220
for i, negative_time in enumerate(negative_times):
    with torch.cuda.amp.autocast(dtype=torch.float16), torch.no_grad():
        out = pipe.alternating(prompt, negative_prompt=negative_prompt, num_inference_steps=30, generator=set_seed(seed),negative_time=negative_time)
        out.images[0].save(f'sunflowers_{i}.png')
        axs[i].imshow(out.images[0])
        axs[i].set_title(titles[i])
        
fig.savefig('sunflowers.png')
    
```
