```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.image as mpimg

# Define the strings to be used for subplot titles and axi labels
str1, str2, str3, str4, str5, str6 = ("an armchair in the shape of an avocado", 
                                       "a boy wearing sunglasses", 
                                       "a painting of a cute corgi wearing a crown",
                                       "cushion in the armchair", 
                                       "a pair of sunglasses with white frame", 
                                       "a crown with red ruby decoration")

str7, str8, str9 = ("Stable Diffusion", "Negative prompt", "Perp.Neg")

image_paths = ['glasses_base_0.png',
               'glasses_trival_0.png',
               'glasses_ours_0.png',
               'glasses_base_0.png',
               'glasses_trival_0.png',
               'glasses_ours_0.png',
               'glasses_base_0.png',
               'glasses_trival_0.png',
               'glasses_ours_0.png']


# Create a figure and a set of subplots
fig, axs = plt.subplots(3, 3, figsize=(15, 15))

# Titles for the left side of the image
left_titles = ['Stable Diffusion', 'Negative prompt', 'Perp.PNG']

# Text for the bottom of the image
bottom_texts = [
    "an armchair in the shape of an avocado\n cushion in the armchair",
    "a boy wearing sunglasses\n a pair of sunglasses with white frame",
    "a painting of a cute corgi wearing a crown\n a crown with red ruby decoration"
]

# Load each image and plot
for i, ax in enumerate(axs.flatten()):
    img = mpimg.imread(image_paths[i])
    ax.imshow(img)
    ax.axis('off')  # Turn off axis
    
    

# Adding text to the left of each row
for idx, ax in enumerate(axs[:, 0]):
    ax.text(-0.5, 0.5, left_titles[idx], verticalalignment='center', horizontalalignment='center',
            transform=ax.transAxes, fontsize=12, rotation=90, fontweight='bold')

# Create boxes for the bottom texts
for idx, ax in enumerate(axs[2, :]):
    # Create a box at the bottom of each column in the last row
    fig.text(idx * 0.3333+ 0.2, 0.05, bottom_texts[idx], ha='center', va='center', fontsize=12,
             bbox=dict(facecolor='white', alpha=0.5))

# Show the plot
plt.show()
```
